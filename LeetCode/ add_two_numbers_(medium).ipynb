```python
# Definition for singly-linked list.
# class ListNode(object):
#     def __init__(self, val=0, next=None):
#         self.val = val
#         self.next = next

class ListNode(object):
    
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next
    
    def __str__(self):
        r = self.__repr__()
        return(r + " val: "+ str(self.val) + " -->\n"+ self.next.__str__())
```

```python
       
class Solution(object):
    
    def getDigits(self, tree, digit_lst):
        """ 
            Get list of digits from a tree of ListNodes
            
                Args:
                    tree - linked list of ListNodes
                    digit_list - list to accamulate digits
                    
                Returns:
                    list with digits                   
        """
        if tree:
            digit_lst.append(tree.val)
            if tree.next:
                return self.getDigits(tree.next, digit_lst)
            else:
                return digit_lst

    
    def getNumber(self, digit_lst):
        """ 
            Convert list of digits to a number
        """
        digits_cnt = len(digit_lst)
        ten_powers = list(range(digits_cnt))
        digits_with_powers_of_ten = zip(digit_lst, ten_powers)
        return sum([x * pow(10,n) for x,n in digits_with_powers_of_ten])
    
    def digits(self, num):
        # Convert number to a list of digits
        return [int(x) for x in str(num)]
    
    @staticmethod
    def digitsListWrk(root, tree, lst):
        # Recursively add list elements to a tree
        # print("tree: ", tree)
        if tree and lst:
            node = ListNode(val=lst[0])
            tree.next = node
            return Solution.digitsListWrk(root, node, lst[1:])
        else:
            return root
    
    @staticmethod
    def digitsList(lst):
        # Add list elements to a tree and return a tree
        root = ListNode()
        if lst:
            root.val = lst[0]
            return Solution.digitsListWrk(root, root, lst[1:])
        else:
            return root # empty tree
        

    def addTwoNumbers(self, l1, l2):
        """
        :type l1: ListNode
        :type l2: ListNode
        :rtype: ListNode
        """
        # Get digit lists from ListNode trees
        digits_lst1 = self.getDigits(l1, [])
        digits_lst2 = self.getDigits(l2, [])
        
        # Get numbers from lists
        num1 = self.getNumber(digits_lst1)
        num2 = self.getNumber(digits_lst2)
        
        # Add numbers and return a tree of ListNodes with digits in reverse order
        result_num = num1 + num2
        result_digits = self.digits(result_num)
        result_digits.reverse()
        print("result digits: ", result_digits)
        result = self.digitsList(result_digits)
        return result
        

```

```python
# Input:
# [2,4,3]
# [5,6,4]
#
# 243 + 564 = 807
#
# Expected:
# [8,0,7]
#
#l1 = ListNode(2, ListNode(4, ListNode(3)))
#l2 = ListNode(5, ListNode(6, ListNode(4)))
l1 = Solution.digitsList([2,4,3])
l2 = Solution.digitsList([5,6,4])
print(Solution().addTwoNumbers(l1,l2))
```
